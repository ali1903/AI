```
#MLP Yapısını kurma (Gereken Kutuphaneleri cagiralim)
from keras.models import Sequential  
from keras.layers import Dense
from sklearn import preprocessing
import numpy
numpy.random.seed(7)

dataset = numpy.loadtxt("diabetes.csv", delimiter=",")

datalar = dataset[:,0:8]
datalar = preprocessing.normalize(datalar)

# : tum satırlar, 0:8 arasi sutunlar (feature)
etiketler = dataset[:,8]
# : tum satırlar, 8.sutun arasi (class)


model = Sequential()
#input_dim=0 cunku 10 feature var
model.add(Dense(100, input_dim=8, activation='relu')) #Hidden layer1
model.add(Dense(140, activation='relu'))              #Hidden layer2
model.add(Dense(220, activation='tanh'))              #Hidden layer2
model.add(Dense(220, activation='relu'))              #Hidden layer2
model.add(Dense(1, activation='sigmoid'))            #Output layer (kac cikis varsa o kadar noron)
#activation=sigmoid, relu, linear, tanh,Elu, softmax, swish softplus, softmax,


model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
#loss=mse, mae, mean_squared_error, cosine_proximity, binary_crossentropy, poisson, KLD
# Modeli verilere gore train etme (Fitting)
model.fit(datalar, etiketler, epochs=400, batch_size=10 )
basarim=model.evaluate(datalar,etiketler)
print("\n %s: %.2f%% " % (model.metrics_names[1], basarim[1]*100)) 
```
