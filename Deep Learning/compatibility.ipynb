```python
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import Conv2D , MaxPooling2D
from keras.layers import Activation , Dropout , Flatten , Dense
from keras import backend as K

img_width, img_height = 224 , 224 
train_data_yolu="Train"
validation_data_yolu="Validation"
train_ornek_sayisi=870
validation_ornek_sayisi=300
epochs=20
batch_size=200
if K.image_data_format()=="channels_first":
    input_shape = (3,img_width , img_height)
else : 
    input_shape = (img_width , img_height , 3)
```

```python
#loss=mse, mae, mean_squared_error, cosine_proximity, binary_crossentropy, poisson, KLD
#activation=sigmoid, relu, linear, tanh,Elu, softmax, swish softplus, softmax,

model=Sequential()
#giriş katmanı ve Conv1
model.add(Conv2D(96,  (11, 11),input_shape=input_shape))
model.add(Activation('softmax'))
model.add(MaxPooling2D(pool_size=(2,2)))
#Conv2
model.add(Conv2D(256,(5,5)))
model.add(Activation("softmax"))
model.add(MaxPooling2D(pool_size=(2,2)))
#Conv3
model.add(Conv2D(384,(3,3)))
model.add(Activation("softmax"))
model.add(MaxPooling2D(pool_size=(2,2)))
#Conv4
model.add(Conv2D(384,(3,3)))
model.add(Conv2D(256,(3,3)))
model.add(Activation("softmax"))
model.add(MaxPooling2D(pool_size=(2,2)))
#Flatte düzleştirme
model.add(Flatten())

#DEnse=#FC (Fully Connected Layer Oluşturuyor)
model.add(Dense(2048)) # 64 nöron var
model.add(Dense(1000)) # 64 nöron var
model.add(Activation("softmax"))
model.add(Dropout(0.5))

#sınıf sayısı
#Binary sınıflama için : model.add(Dense(4))
model.add(Dense(4)) # 64 nöron var
model.add(Activation("sigmoid"))
#model derleme
model.compile(loss="binary_crossentropy",optimizer="RMSprop",metrics=["accuracy"])
```

```python
#Train işlemi için kullanılan verileri cogaltıyor
train_datalar = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
    )
#datalar = ImageDataGenerator(rotation_range=40,width_shift_range=0.2,height_shift_range=0.2,shear_range=0.2,zoom_range =0.2,horizontal_flip=True,fill_node="nearest")
#test işlemi için görüntüleri rescale boyutlandırma yapıyor
test_datalar=ImageDataGenerator(rescale=1. /255)

train_generator = train_datalar.flow_from_directory(
    train_data_yolu,
    target_size=(img_width,img_height),
    batch_size=batch_size,
    class_mode='categorical')


validation_generator = test_datalar.flow_from_directory(
    validation_data_yolu,
    target_size=(img_width,img_height),
    batch_size=batch_size,
    class_mode="categorical")
```

```python
#modelin Eğitimi
model.fit(
    train_generator,
    steps_per_epoch=train_ornek_sayisi // batch_size,
    epochs=epochs,
    validation_data=validation_generator,
    validation_steps=validation_ornek_sayisi // batch_size)
```
