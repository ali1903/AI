**Verileri Oluştur**

```
#Gereken kutuphaneler yuklenmeli
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split  #Rastgele Train Test Bolmek icin
import numpy

numpy.random.seed(7)
#veriseti yukleme
dataset=numpy.loadtxt("diabetes.csv", delimiter=",")

#datalar 0:8 arasi featurelar alindi
datalar=dataset[:,0:8]
#etiketler 8.indis class (outcome) alindi
etiketler=dataset[:,8]

# Splitting data
X_train, X_test, Y_train, Y_test = train_test_split(datalar, etiketler, test_size=0.2, random_state=0)
#X_train: Train Datalari
#X_test:  Test Dataları
#Y_train: Train Etiketleri
#Y_test:  Test Etiketleri
```

**MLP Yapisi Kurulumu**

```
model=Sequential()
#8 feature girisli oldugu icin input_dim=8 secildi
model.add(Dense(25, input_dim=8, activation='relu')) # Hidden Layer 1 (25 noron)
#hidden layer 2 ye input_dim yazilmaz.
model.add(Dense(50, activation='tanh'))              # Hidden Layer 2  (50 noron)
#Cikis diabet-nondiabet oldugu icin tek cikisla ifade edilebilir
model.add(Dense(1, activation='sigmoid'))            #Output Layer
#activation=sigmoid, relu, linear, tanh

#Modeli Derleme
model.compile(loss='KLD', optimizer='adam', metrics=['accuracy'])
#loss=KLD, mae, mse, mean_squared_error, cosine_proximity, binary_crossentropy, poisson

#batch_size: Egitim sirasinda verisetinden ornekler kacar kacar alinacak
model.fit(X_train, Y_train, epochs=100, batch_size=10)
basarim=model.evaluate(X_train, Y_train)

print("\n Train %s : %.2f%% " % (model.metrics_names[1], basarim[1]*100))
```
