**Kutuphaneler Yuklensin**

```
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics.classification import accuracy_score
from sklearn.neural_network import MLPClassifier
```

**Verileri Yukle ve Test-Train Olarak Ayir**

```
np.random.seed(7)
#veriseti yukleme
dataset=np.loadtxt("diabetes.csv", delimiter=",")

#datalar 0:8 arasi featurelar alindi
datalar=dataset[:,0:8]
#etiketler 8.indis class (outcome) alindi
etiketler=dataset[:,8]

# Splitting data
X_train, X_test, Y_train, Y_test = train_test_split(datalar, etiketler, test_size=0.2, random_state=0)
```

```
mlp = MLPClassifier(random_state=1, hidden_layer_sizes=(10, 10, 10), 
                    learning_rate='adaptive',
                    solver='adam', 
                    max_iter=300).fit(X_train, Y_train)
# MLPClassifier parametreleri
# alphafloat, default=0.0001 : L2 penalty (regularization term) parameter
# learning_rate{‘constant’, ‘invscaling’, ‘adaptive’}, default=’constant’
# solver=‘lbfgs’, ‘sgd’, ‘adam’, default=’adam’
# learning_rate_init, default=0.001
# epsilon, default=1e-8

#Y_pred = mlp.predict(X_test)
#acc = accuracy_score(Y_test, Y_pred)

acc=mlp.score(X_test, Y_test)
print('Done.\nAccuracy: %f' % acc)


```

**Confusion Matrix**

```
from sklearn.metrics import classification_report, confusion_matrix
Y_pred = mlp.predict(X_test)
cm = confusion_matrix(Y_pred, Y_test)

print(cm)
print(classification_report(Y_test,Y_pred))
```
